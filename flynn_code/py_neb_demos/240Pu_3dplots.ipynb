```python
import numpy as np

import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
path = np.loadtxt('./Eric_240Pu_LAP_Mass_True_dt_.5_path.txt',delimiter=',',skiprows=1)
git_path = np.loadtxt('../../Paths/240Pu/Eric_240Pu_LAP_Mass_True_path.txt',delimiter=',',skiprows=1)
```

```python
%matplotlib notebook
fig = plt.figure()
ax = plt.axes(projection ='3d')

ax.plot3D(path[:,0], path[:,1], path[:,2], 'green',label='new')
ax.plot3D(git_path[:,0], git_path[:,1], git_path[:,2], 'red',label='git')
plt.legend()
plt.show()
```
